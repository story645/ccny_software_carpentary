# Conditional Indexing #

```python
#boolean indexing
train['Age'][train['Age']>60] #take every row of the age column where the age is greater than 60
```

```python
train['Name'][train['Age']>60] #the name of every person over 60
```

```python
train['Name'][train['Age']!=60] #everyone who's not 60
```

```python
train['Name'][~(train['Age']!=60)]#negate the condition age!=60, so everyone who is 60
```

```python
#combine conditions
#everyone between 60 and 75
train[(train['Age']>60)&(train['Age']<75)]
```

```python
#lets look at extreme populations: under 2 or over 75

train[(train['Age']>75)|(train['Age']<2)]
```

# Grouping #

```python
#order the train data set by splitting by class
by_pclass = train.groupby('Pclass')
```

```python
by_pclass #is a groupby object, needs to be further processed to be used
```

```python
by_pclass.count() #how many of each record
```

```python
by_pclass.size()
```

```python
#lets calculate the survival rate by class
# Survived is coded 0 = death, 1 = lived
# Count counts every element that's not NaN
Survival_rate = by_pclass['Survived'].sum()/by_pclass['Survived'].count()
print Survival_rate
```

```python
#shorter way to complete the survival rate
#survival = (sum of codes)/ (number of people in class) = sum/count
by_pclass['Survived'].mean()
```

```python
by_pclass['Age'].mean()
```

```python
by_pclass['Fare'].max()
```

```python
by_pclass['Fare'].min()
```

```python
by_pclass['Fare'].describe() #all summary stats
```

```python
#splits the data into class and then by sex
#then computes the survival rate by subgroup 
train.groupby(['Pclass', 'Sex'])['Survived'].mean()
```

```python
#prettify the table by rearranging by groups
train.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()
```

```python
#create a pivot table to do advanced grouping
#index is what becomes the row
#agg function is how the data in each subset->(Pclass, Sex) is analyzed
#first argument is the name of the column being analyzed
train.pivot_table('Survived', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
#do multiple aggregations
#A python dictionary is {key:value} and usually used as a data structure
#instead of specifying one value to analyze at the beggining, multiple can be defined in the agg function
#the form is {column:aggregation_function}
train.pivot_table(index='Pclass', columns='Sex', aggfunc={'Survived':'sum', 'Fare':'mean'})
```

```python
#take the class by gender table and assign it to a variable
pivot_pclass_sex = train.pivot_table('Survived', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
import matplotlib.pyplot as plt #rename matplotlib.pyplot as plt
from matplotlib import style #make the graph prettier
%matplotlib inline

#plot the table
pivot_pclass_sex.plot(kind='bar')
```
